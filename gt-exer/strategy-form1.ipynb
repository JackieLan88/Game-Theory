```python
# Import Pygambit and create a game object
# prisoner's dilemma is a strategic form game, it can be created in 
# a tabular fashion with Game.new_table

import numpy as np 
import pygambit as gbt

n_strategies = [2,2]
g = gbt.Game.new_table(n_strategies, title="Prisoner's Dilemma")

print(type(g))

# let's name the players and each of their possible strategies, in both cases
# "cooperate" and "defect"  (though not necessary but recommended)

g.players[0].label = "Tom"
g.players[0].strategies[0].label = "Cooperate"
g.players[0].strategies[1].label = "Defect"

g.players[1].label = "Jerry"
g.players[1].strategies[0].label = "Cooperate"
g.players[1].strategies[1].label = "Defect"

# next step: let's assign payoffs for each of the game's possible outcomes:

# both cooperate
g["Cooperate","Cooperate"]["Tom"] = -1
g["Cooperate","Cooperate"]["Jerry"] = -1

# Tom cooperate, Jerry defects
g["Cooperate","Defect"]["Tom"] = -3
g["Cooperate","Defect"]["Jerry"] = 0

# Tom defects, Jerry cooperates
g["Defect","Cooperate"]["Tom"] = 0
g["Defect","Cooperate"]["Jerry"] = -3

# both defect

g["Defect","Defect"]["Tom"] = -2
g["Defect","Defect"]["Jerry"] = -2

g   # --- this payout matrix structure is described in Game theory as 
            # --- the strategic form or normal form representation of a game
            


```

# Creating games from arrays

```python
# the most direct way to create a strategiv form game is via Game.from_arrays()

# one n-dimensional array per player
# n is the number of players in the game

player1_payoff = np.array([[-1,-3],[0,-2]])
player2_payoff = np.transpose(player1_payoff)

g1 = gbt.Game.from_arrays(
    player1_payoff,
    player2_payoff,
    title= "Another Prisoner's Dilemma"

)

g1
```
